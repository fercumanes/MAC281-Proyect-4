```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
G.add_nodes_from(["Beverly", "Andre","Carol","Diane","Ed","Fernando","Garth","Heather","Ike","Jane"])
G.add_edges_from([("Beverly","Andre"),("Beverly","Ed"),("Beverly","Garth"),("Beverly","Diane")])
G.add_edges_from([("Andre","Diane"),("Andre","Carol"),("Andre","Fernando")])
G.add_edges_from([("Ed","Diane"),("Ed","Garth")])
G.add_edges_from([("Diane","Carol"),("Diane","Fernando"),("Diane","Garth")])
G.add_edge("Carol","Fernando")
G.add_edges_from([("Garth","Fernando"),("Garth","Heather")])
G.add_edge("Fernando","Heather")
G.add_edge("Heather","Ike")
G.add_edge("Ike","Jane")
nx.draw(G, with_labels=True, font_color='blue')
plt.show()
```
